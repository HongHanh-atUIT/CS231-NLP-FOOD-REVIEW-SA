```python
import json
import re
```

```python
def normalize_name(description: str) -> str:
    """
    Chuyển mô tả emoji về dạng key:
    "waving hand: light skin tone" → "emj_waving_hand"
    """
    base = description.split(":")[0].strip().lower()
    base = re.sub(r"[^a-z0-9]+", "_", base).strip("_")
    return f"emj_{base}"

def load_emoji_entries(paths):
    """Đọc tất cả file JSON và yield từng hàng hợp lệ."""
    for path in paths:
        with open(path, "r", encoding="utf-8") as f:
            for row in json.load(f):
                if len(row) == 4:
                    yield row  # Dạng: [id, code, emoji, desc]

def build_mapping(EMOJI_FILES):
    """Trả về danh sách các dòng dạng: emoji<TAB>normalized_name"""
    lines = []
    for _, _, emoji, desc in load_emoji_entries(EMOJI_FILES):
        lines.append(f"{emoji}\t{normalize_name(desc)}")
    return lines

def save_txt(lines, output_path="Dataset/emoji_dict.txt"):
    """Ghi danh sách dòng vào file TXT."""
    with open(output_path, "w", encoding="utf-8") as f:
        f.write("\n".join(lines))
    print(f"Đã tạo {output_path} thành công!")
```

```python
# File nguồn
EMOJI_FILES = [
    "Dataset/all-emoji.json",
    "Dataset/full-emoji-modifiers.json",
]

save_txt(build_mapping(EMOJI_FILES))
```
